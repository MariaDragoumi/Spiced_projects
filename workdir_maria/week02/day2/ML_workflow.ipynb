```python
#Normalization
#output range is [0,1]
#does not deal well with outliers
```

```python
def normalize(X):
    return (X-X.min())/(X.max()-X.min())
```

```python
#Standardization
#will be centered around zero and go about 3 (3 std have 99% of data)
# deals well wit outliers
# scikit-learn equivalent: StandardScaler()
```

```python
def standardize(X):
    return (X-X.mean())/X.std()
```
